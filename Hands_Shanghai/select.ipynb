```python
import pandas as pd
import os
```

```python
def take_int(filename):
    name = filename.split('_')[0]
    return int(name)

filepath = '2500_con_res'
files = os.listdir(filepath)
files.sort(key = take_int)

for file in files:
    if file.endswith('.xlsx'):
        try:
            df = pd.read_excel(filepath + '/' + file)
            context = df.iloc[:20, 2]

            for text in context:
                text = text.replace(' ', '')
                if '采购' in text or '销售' in text:
                    name = file.split('_')[0]
                    print(f'采购文件：{name} \n')
                    break
        except Exception as e:
            print(f"发生错误：{e}")

```
